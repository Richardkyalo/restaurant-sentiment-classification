```python
#sentiment analysis using vader approach
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
from tqdm.notebook import tqdm
sia= SentimentIntensityAnalyzer()
```

```python
sia.polarity_scores(example)
df
```

```python
#run the polarity sore on the entire data set
compound_scores = []
positive_scores = []
negative_scores = []
neutral_scores = []
for index, row in tqdm(df.iterrows(), total=len(df)):
    text=row['Review']
    sentiment_scores = sia.polarity_scores(text)

    # Print sentiment scores
    # print(f"Sentiment scores for text {index + 1}:")
    # print(f"Compound: {sentiment_scores['compound']:.2f}")
    # print(f"Positive: {sentiment_scores['pos']:.2f}")
    # print(f"Negative: {sentiment_scores['neg']:.2f}")
    # print(f"Neutral: {sentiment_scores['neu']:.2f}")
    # print()

    # Store sentiment scores
    compound_scores.append(sentiment_scores['compound'])
    positive_scores.append(sentiment_scores['pos'])
    negative_scores.append(sentiment_scores['neg'])
    neutral_scores.append(sentiment_scores['neu'])

df['compound_score'] = compound_scores
df['positive_score'] = positive_scores
df['negative_score'] = negative_scores
df['neutral_score'] = neutral_scores
```

```python
#pretrained model
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = "cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL, force_download=True)
```

```python
# VADER results
sia.polarity_scores(example)
```

```python
# roberta results

encoded_text=tokenizer(example, return_tensors='pt')
output=model(**encoded_text)
scores=output[0][0].detach().numpy()
scores= softmax(scores)
scores_dict={
    "roberta neg": scores[0],
    "roberta neu": scores[1],
    "roberta pos": scores[2]
}
```

```python
scores_dict
```

```python
def polarity_scores_roberta(example):
    encoded_text=tokenizer(example, return_tensors='pt')
    output=model(**encoded_text)
    scores=output[0][0].detach().numpy()
    scores= softmax(scores)
   
    return scores

polarity_scores_roberta(example)
    
```

```python
sentiment_scores_dict = {
    "compound_score": [],
    "positive_score": [],
    "negative_score": [],
    "neutral_score": []
}
sentiment_scores_roberta_dict={
    "roberta neg": [],
    "roberta neu": [],
    "roberta pos": []
}
for index, row in tqdm(df.iterrows(), total=len(df)):
    index+=1
    try:
        
        text=row['Review']
        sentiment_scores = sia.polarity_scores(text)
    
        sentiment_scores_roberta=polarity_scores_roberta(text)
    
        sentiment_scores_dict['compound_score'].append(sentiment_scores['compound'])
        sentiment_scores_dict['positive_score'].append(sentiment_scores['pos'])
        sentiment_scores_dict['negative_score'].append(sentiment_scores['neg'])
        sentiment_scores_dict['neutral_score'].append(sentiment_scores['neu'])
    
        sentiment_scores_roberta_dict['roberta neg'].append(sentiment_scores_roberta[0])
        sentiment_scores_roberta_dict['roberta neu'].append(sentiment_scores_roberta[1])
        sentiment_scores_roberta_dict['roberta pos'].append(sentiment_scores_roberta[2])
    
        sentiment_scores_roberta_dict['roberta neg'] = [float(x) for x in sentiment_scores_roberta_dict['roberta neg']]
        sentiment_scores_roberta_dict['roberta neu'] = [float(x) for x in sentiment_scores_roberta_dict['roberta neu']]
        sentiment_scores_roberta_dict['roberta pos'] = [float(x) for x in sentiment_scores_roberta_dict['roberta pos']]
        
    except RuntimeError:
        print(f'broke for id', index)


combined_sentiment_scores_dict = {**sentiment_scores_dict, **sentiment_scores_roberta_dict}

print(combined_sentiment_scores_dict)
```

```python
for key, value in combined_sentiment_scores_dict.items():
    if len(value) == len(df):
        df[key] = value
    else:
        df[key] = np.nan
        min_len = min(len(value), len(df))
        df.loc[:min_len-1, key] = value[:min_len]

print(df.head())
```
